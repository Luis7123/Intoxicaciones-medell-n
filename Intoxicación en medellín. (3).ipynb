* Veamos ahora la cantidad de hombres y mujeres intoxicados cada año, visualizaremos el aumento o disminución de los casos para cada año:

```python
years = []; years_r = []; sexo = []; cantidad = []

for y in file['year']:
    if years_r.count(y) < 2:
        years_r.append(y)
    if y not in years:
        years.append(y)
        sexo.append('M')
        sexo.append('F')

for y1 in years:
    cantidad.append(len(file[(file['year'] == y1) & (file['sexo'] == 'M')]))
    cantidad.append(len(file[(file['year'] == y1) & (file['sexo'] == 'F')]))
```

```python
dataMF = {'year':years_r, 'sexo':sexo, 'cantidad':cantidad}
df_dataMF = pd.DataFrame(dataMF)
df_dataMF
```

```python
sns.lineplot(data=df_dataMF, x="year", y="cantidad", hue='sexo'); plt.title('cantidad de intoxicados por año')
```

Vemos que los registros de intoxicaciones presentan un aumento respecto a los años anteriores, teniendo únicamente un decrecimiento considerable hacia el año 2013.

# Intoxicaciones por comuna. 

```python
comunas = []; comunas_sr = []

for c in file['comuna']:
    if comunas.count(c) < 2:
        comunas.append(c)
    if c not in comunas_sr:
        comunas_sr.append(c)

comunas.sort(); comunas_sr.sort()

sexo_comunas = []; cantidad_comunas = []

for c in comunas_sr:
    sexo_comunas.append('M')
    sexo_comunas.append('F')
    cantidad_comunas.append(len(file[(file['comuna'] == c) & (file['sexo'] == 'M')]))
    cantidad_comunas.append(len(file[(file['comuna'] == c) & (file['sexo'] == 'F')]))

intoxicados_comunas_generos = pd.DataFrame({'comuna':comunas, 'sexo':sexo_comunas, 'cantidad':cantidad_comunas})

plt.figure(figsize=(11,5))
graf = sns.barplot(x="comuna", y="cantidad", hue="sexo", data=intoxicados_comunas_generos)
graf.set_xticklabels(graf.get_xticklabels(), rotation=40, ha="right")
plt.title('intoxicaciones por comuna', size=14)
#plt.savefig('comunas.png', dpi=300, bbox_inches='tight')  # Ejecutar para guardar
plt.show()
```
